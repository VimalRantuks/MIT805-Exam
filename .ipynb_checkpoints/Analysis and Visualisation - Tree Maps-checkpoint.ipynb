```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import squarify
import seaborn as sns
%matplotlib inline
```

```python
# Import map reduce poutput csv
df = pd.read_csv("Liquor_sales_reduced_items.csv", index_col=0)
df
```

```python
# change the date format of recent date column
df["Recent_Date"] = pd.to_datetime(df["Recent_Date"],format = "%Y/%m/%d")
```

```python
df = df[df["Recent_Date"]>"2018-01-01"]
df
```

```python
# Since we have 2 months of data, we choose 2 days after the end of February as a time stamp.The last
# purchase date made per user is subtracted from the time stamp to get the recency value.
import datetime as dt
study_date = dt.datetime(2020,3,2)
df["Recency"] = study_date - df["Recent_Date"]
df['Recency'].astype('timedelta64[D]')
df['Recency']=df['Recency'] / np.timedelta64(1, 'D')
```

```python
df
```

```python
finaldf = df[["user_id","Monetary","Frequency","Recency"]]
```

```python
finaldf
```

```python
# Create labels for Recency, Monetary and Frequency and assign quantiles
rec_labels = range(4,0,-1)
fr_labels = range(1, 5)
mon_labels = range(1, 5)
r_groups = pd.qcut(finaldf['Recency'], q=4, labels=rec_labels)
f_groups = pd.qcut(finaldf['Frequency'], q=4, labels=fr_labels)
m_groups = pd.qcut(finaldf['Monetary'], q=4, labels=mon_labels)

finaldf = finaldf.assign(R = r_groups.values, F = f_groups.values,M = m_groups.values)
finaldf.head()
```

```python
# determine rfm product score by multiplying the category labels
finaldf[["R", "F","M"]] = finaldf[["R", "F","M"]].apply(pd.to_numeric)
# Calculate RFM_Score
finaldf['RFM_Product'] = finaldf["R"]*finaldf["F"]*finaldf["M"]
print(finaldf['RFM_Product'].head())
```

```python
finaldf["RFM_Product"].describe()
```

```python
#Determine the rfm groups based on scaling
def rfmgroup(df):
    if df['RFM_Product'] >= 48:
        return 'Key Important Customers'
    elif ((df['RFM_Product'] >=36) and (df['RFM_Product'] < 48)):
        return 'Great Customers'
    elif ((df['RFM_Product'] >= 24) and (df['RFM_Product'] < 36)):
        return 'Loyal'
    elif ((df['RFM_Product'] >= 12) and (df['RFM_Product'] < 24)):
        return 'Good Potential'
    else:
        return 'Focus for Retention'

finaldf['RFM_Level'] = finaldf.apply(rfmgroup, axis=1)
finaldf.head()
```

```python
# Calculate and return a size of each segment 
rfm_agg = finaldf.groupby('RFM_Level').agg({
    'RFM_Level':'count'
}).round(1)
print(rfm_agg)
```

```python
# Use of squarify to plot a tree map of segments
import squarify

rfm_agg.columns = ['RFM_Level']
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 9)
squarify.plot(sizes=rfm_agg['RFM_Level'], 
              label=['Focus for Retention',
                     'Good Potential',
                     'Great Customers',
                     'Key Important Customers',
                     'Loyal'],
              color=["red",'yellow','blue', "green",'grey'],
              alpha=.6 )
plt.title("RFM Segmentation",fontsize=25,fontweight="bold")
plt.axis('off')
plt.show()
```
